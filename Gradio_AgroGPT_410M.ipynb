```python
import gradio as gr

from transformers import pipeline

pipe = pipeline("text-generation", model="f0rGoTTen000/AgroGPT_125M")
```

```python
def predict(text):
 return pipe(text)[0]["generated_text"]

demo = gr.Interface(
  fn=predict,
  inputs='text',
  outputs='text',
)

demo.launch()
```
